```python
import spacy
import pandas as pd
import PyPDF2
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
nlp = spacy.load("en_core_web_sm")

```

```python
jobs = {
    "Data Scientist": "python machine learning pandas numpy statistics",
    "Web Developer": "html css javascript react nextjs",
    "AI Engineer": "deep learning tensorflow pytorch nlp",
    "Cyber Security Analyst": "network security ethical hacking firewall",
    "Cloud Engineer": "aws azure cloud docker kubernetes"
}

job_df = pd.DataFrame(list(jobs.items()), columns=["Job", "Skills"])
job_df
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
file_name = list(uploaded.keys())[0]

def read_resume(file):
    pdf = PyPDF2.PdfReader(file)
    text = ""
    for page in pdf.pages:
        text += page.extract_text()
    return text.lower()

resume_text = read_resume(open(file_name,'rb'))
resume_text[:500]
```

```python
def extract_skills(text):
    doc = nlp(text)
    skills = []
    for token in doc:
        if token.pos_ in ["NOUN", "PROPN"]:
            skills.append(token.text)
    return " ".join(set(skills))

resume_skills = extract_skills(resume_text)
resume_skills[:300]
```

```python
def recommend_job(resume_skills):
    documents = job_df["Skills"].tolist() + [resume_skills]
    tfidf = TfidfVectorizer()
    vectors = tfidf.fit_transform(documents)
    similarities = cosine_similarity(vectors[-1], vectors[:-1])
    job_df["Match Score"] = similarities[0]
    return job_df.sort_values(by="Match Score", ascending=False)
```

```python
final_result = recommend_job(resume_skills)
final_result
```
